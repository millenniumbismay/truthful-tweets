```python
from googlesearch import search
from bs4 import BeautifulSoup
import requests

def get_top_articles(query, num_articles=5):
    articles = []

    # Perform Google search
    search_results = search(query, num_results=num_articles)

    # Iterate over search results
    for result in search_results:
        try:
            # Fetch the HTML content of the search result page
            response = requests.get(result)
            soup = BeautifulSoup(response.text, 'html.parser')

            # Extract article title and content
            title = soup.title.string
            content = soup.get_text()

            # Append article information to the list
            articles.append({'title': title, 'content': content, 'url': result})
        except Exception as e:
            print(f"Error processing URL: {result} - {e}")

    return articles

# Example: Get top 5 articles for a list of topics
topics = ["Dan Patrick, drive-in voting, people of color, statistics, PolitiFact ruling"]
for topic in topics:
    print(f"\nTop 5 articles for '{topic}':\n")
    articles = get_top_articles(topic, num_articles=5)

    for i, article in enumerate(articles, start=1):
        print(f"{i}. Title: {article['title']}")
        print(f"   URL: {article['url']}")
        print(f"   Content:\n{article['content'][:1000]}...\n")  # Displaying the first 300 characters of the content for brevity
```
