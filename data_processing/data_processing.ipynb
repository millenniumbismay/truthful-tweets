```python
import re
import string

def preprocess_tweets(tweet):
  ### Remove URLs
  tweet = re.sub(r'http\S+', '', tweet)
  ### Removing any string with .com
  tweet = re.sub(r'\b\w+\.com\b', '', tweet)
  ### Remove usernames and words containing @ - generally emails
  tweet = re.sub(r'[^\s]*@[^\s]*', '', tweet)
  ### Remove punctuations
  tweet = tweet.translate(str.maketrans('', '', string.punctuation))
  ### Removing \n
  tweet = re.sub('\n', ' ', tweet)
  ### Correcting word lengthening - removing more than 2 consecutive repitations of characters
  tweet_word_list = tweet.split(' ')
  pattern = re.compile(r"(.)\1{2,}")
  tweet = ' '.join(pattern.sub(r"\1\1", word) for word in tweet_word_list)
  return tweet.strip()
```

```python
final_df['cleaned_tweet'] = final_df.apply(lambda x: preprocess_tweets(x['tweet']), axis = 1)
```

```python
final_df_3_majority = final_df.drop(['5_label_majority_answer'], axis = 1)
final_df_3_majority.head()
```

```python
print(final_df_3_majority.shape)
final_df_3_majority = final_df_3_majority[final_df_3_majority['3_label_majority_answer']!='NO MAJORITY']
print(final_df_3_majority.shape)
```

```python
def generateTruthfulness(target, majority_answer):
  if target == True and majority_answer == 'Agree':
    return True
  elif target == True and majority_answer == 'Disagree':
    return False
  elif target == False and majority_answer == 'Agree':
    return False
  elif target == False and majority_answer == 'Disagree':
    return True


final_df_3_majority['Truthfulness'] = final_df_3_majority.apply(lambda x: generateTruthfulness(x['target'], x['3_label_majority_answer']), axis = 1)
final_df_3_majority.head()
```

```python
final_df_5_majority = final_df.drop(['3_label_majority_answer'], axis = 1)
final_df_5_majority.head()
```

```python
print(final_df_5_majority.shape)
final_df_5_majority = final_df_5_majority[final_df_5_majority['5_label_majority_answer']!='NO MAJORITY']
print(final_df_5_majority.shape)
```

```python
def generateTruthfulness(target, majority_answer):
  if target == True and majority_answer == 'Agree':
    return 'True'
  elif target == True and majority_answer == 'Disagree':
    return 'False'
  elif target == True and majority_answer == 'Mostly Agree':
    return 'Mostly True'
  elif target == True and majority_answer == 'Mostly Disagree':
    return 'Mostly False'
  elif target == False and majority_answer == 'Agree':
    return 'False'
  elif target == False and majority_answer == 'Disagree':
    return 'True'
  elif target == False and majority_answer == 'Mostly Agree':
    return 'Mostly False'
  elif target == False and majority_answer == 'Mostly Disagree':
    return 'Mostly True'


final_df_5_majority['Truthfulness'] = final_df_5_majority.apply(lambda x: generateTruthfulness(x['target'], x['5_label_majority_answer']), axis = 1)
final_df_5_majority.head()
```
